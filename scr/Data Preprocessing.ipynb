```r
# Define a function to clean sickday variables
clean_sickday <- function(x) {
  # Replace "Absent" with "absent"
  x <- gsub("Absent", "absent", x, fixed = TRUE)
  # Replace "<U=00A0>" and "<U+00A0>" with NA
  x <- gsub("<U=00A0>|<U+00A0>", NA, x, fixed = TRUE)
  # Replace "doppelt MZpresent117" with NA
  x <- gsub("doppelt MZpresent117", NA, x, fixed = TRUE)
  # Replace "N" with NA
  x <- gsub("N", NA, x, fixed = TRUE)
  # Replace "presentN" with NA
  x <- gsub("presentN", NA, x, fixed = TRUE)
  # Replace "Absent presentUSGEAbsent<U+00DC>LLT" with NA
  x <- gsub("Absent presentUSGEAbsent<U\\+00DC>LLT", NA, x, fixed = TRUE)
  # Remove other characters and keep only "present", "absent", "sick", and NA
  x <- gsub("[^present|absent|sick]", NA, x)
  return(x)
}
```

```r
# Apply the cleaning function to all sickday variables
sickdays[, 4:ncol(sickdays)] <- lapply(sickdays[, 4:ncol(sickdays)], clean_sickday)
```

```r
# Check unique values for all variables
unique_values <- sapply(sickdays, unique)

# Print the unique values for each variable
print(unique_values)
```
